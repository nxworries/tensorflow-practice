## Steps in modelling with tensorflow

1. create a model - define input and output layers, as well as hidden layers of a deep learning model

1. compiling a model - define the loss function (like the cost function), the optimizer(tells the model how to improve), and evaluation metrics

1. fitting the model

````python
# set random seed
tf.random.set_seed(42)

# 1: create model using sequential API
model = tf.keras.Sequential([
        tf.keras.layers.Dense(1)  # one single hidden layer with 1 neuron                
])

# 2: compile
model.compile(loss=tf.keras.losses.mae,  # mean absolute error
              optimizer = tf.keras.optimizers.SGD(), # stochastic grad desc
              metrics=["mae"])
# 3: fit the model
model.fit(X,Y, epochs = 100)
````

````python
# predict using our model
model.predict([0.0])
# pretty bad
````

## Improving the model 

We can improve the model by altering the steps we made to create the model

1. **creating the model** - add layeys, increase number of neurons within the layers, change the activation function

1. **compiling** - can change the optimization function or the learning rate of the optimization function

1. **fitting the model** - change epochs, training data, ets


## A much larger model

````python
# just as an example, not a good idea
# step 1 - creating the model
model2 = tf.keras.Sequential([
       tf.keras.layers.Dense(100, activation = "relu"), 
       tf.keras.layers.Dense(100, activation = "relu"), 
       tf.keras.layers.Dense(100, activation = "relu"),
       tf.keras.layers.Dense(1) 
])
# step 2 - compile
model2.compile(loss = tf.keras.losses.mae,
               optimizer = tf.keras.optimizers.Adam(lr=.0001), 
               metrics=["mae"])
# step 3 - train/fit
model2.fit(X, Y, epochs=100)
# its actually shit! way over complicated
````

````python
# back to model 1
# set random seed
tf.random.set_seed(42)

# 1: create model using sequential API
model = tf.keras.Sequential([
        tf.keras.layers.Dense(20),
        tf.keras.layers.Dense(1)  # one single hidden layer with 1 neuron                
])

# 2: compile
model.compile(loss=tf.keras.losses.mae,  # mean absolute error
              optimizer = tf.keras.optimizers.Adam(), # stochastic grad desc
              metrics=["mae"])
# 3: fit the model
model.fit(X,Y, epochs = 100)
````

````python
model.predict([17.0]).astype(float)
````

````python
# Dan's improvements
# step 1
model = tf.keras.Sequential([
         tf.keras.layers.Dense(50, activation = "relu"),
         tf.keras.layers.Dense(1)                    
])

# step 2
model.compile(loss = 'mae',
              optimizer=tf.keras.optimizers.Adam(lr = .01),
              metrics = ['mae'])
# step 3
model.fit(X, Y, epochs =100)
              
````

### Visualizing model predictions

To visualize predictions, its a good idea to plot them against the ground truth labels.

Often you will see this in the form of y_test or y_true vs y_pred
i.e testing data response vs predicted response

````python
y_pred = model.predict(X_test)
y_pred, Y_test
````

````python
def plot_predictions(
    train_data = X_train,
    train_labels = Y_train,
    test_data = X_test,
    test_labels = Y_test,
    predictions = y_pred):
  ''' 
  plots training data, test data, and compares predictions to ground truth
  '''
  plt.figure(figsize = (10, 7))
  plt.scatter(train_data, train_labels, c = 'b', label = 'training data')
  plt.scatter(test_data, test_labels, c='g', label = 'testing data')
  plt.scatter(test_data, predictions, c='r', label = 'predictions')
  plt.legend
````

````python
plot_predictions()
````

### Evaluating our models prediction with regression evaluation metrics

* MAE - mean absolute error
* MSE - yup
* Huber, R-squared, .....

````python
#Eval the model on the test set
model.evaluate(X_test, Y_test)
````

````python
#manually calc mse and mae
tf.keras.losses.MAE(y_true=Y_test, y_pred=tf.squeeze(y_pred))
# squeezing the y_pred tensor to make it the same dimention as y_test
````

````python
tf.keras.losses.MSE(y_true=Y_test, y_pred=tf.squeeze(y_pred))
````

````python
# make some functions to reuse MAE and MSE
def mae(y_true= Y_test, y_pred=tf.squeeze(y_pred)):
  return tf.keras.losses.MAE(y_true=tf.squeeze(y_true), y_pred=tf.squeeze(y_pred))
def mse(y_true=Y_test, y_pred=y_pred):
  return tf.keras.losses.MSE(y_true=tf.squeeze(y_true), y_pred=tf.squeeze(y_pred))
````

### Improving the Model!!!!!!!

- experiment, experiment, experiment

more/better quality data :\, make model larger/more complex, train for longer, ... 

3 modelling experiments that will follow:

1. `model_1` - same as the og model - 1 layer, trained for 100 epochs

1. `model_2` - 2 layers trained for 100 epochs, Adam not SGD

1. `model_3` - 2 layers trained for 500 epochs, Adam

1. `model_4` - model 3 with another layer

1. `model_5` - model 3 with higher learning rate and smaller number of epochs


````python
# set random seed
tf.random.set_seed(42)

model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model_1.compile(loss=tf.keras.losses.mae,
                optimizer = tf.keras.optimizers.SGD(),
                metrics = ['mae'])
model_1.fit(X_train, Y_train, epochs = 100)


````

````python
# make + plot predictions
y_preds_1 = model_1.predict(X_test)
plot_predictions(predictions=y_preds_1)
````

````python
# calculate evaluation metrics for mod 1
mae_1 = mae(Y_test, y_preds_1)
mse_1 = mse(Y_test, y_preds_1)
print(mse_1, mae_1)
````

````python
# model 2
model_2 = tf.keras.Sequential([
         tf.keras.layers.Dense(12),
         tf.keras.layers.Dense(1)
])
model_2.compile(loss = tf.keras.losses.mae,
                optimizer = tf.keras.optimizers.Adam(learning_rate=.015),
                metrics = ['mae'])
model_2.fit(X_train, Y_train, epochs = 100);
````

````python
y_preds_2 = model_2.predict(X_test)

plot_predictions(predictions=y_preds_2)

````

````python
mae_2 = mae(Y_test, y_preds_2)
mse_2 = mse(Y_test, y_preds_2)
mae_2, mse_2
````

````python
#create model
model_3 = tf.keras.Sequential([ 
            tf.keras.layers.Dense(12),
            tf.keras.layers.Dense(1)
])
# compile model
model_3.compile(loss = tf.keras.losses.mae,
                optimizer = tf.keras.optimizers.Adam(learning_rate=.015),
                metrics = ['mae'])
model_3.fit(X_train, Y_train, epochs = 500)
````

````python
y_preds_3 = model_3.predict(X_test)
plot_predictions(predictions = y_preds_3)
````

````python
mae_3 = mae(y_true = Y_test, y_pred=y_preds_3)
mse_3 = mse(y_true = Y_test, y_pred=y_preds_3)
mae_3, mse_3
````

````python
model_4 = tf.keras.Sequential([
            tf.keras.layers.Dense(10),
            tf.keras.layers.Dense(10),
            tf.keras.layers.Dense(1)
])
model_4.compile(loss = tf.keras.losses.mae,
                optimizer = tf.keras.optimizers.Adam(learning_rate=.015),
                metrics = ['mae'])
model_4.fit(X_train, Y_train, epochs = 500)
````

````python
y_preds_4 = model_4.predict(X_test)
plot_predictions(predictions=y_preds_4) 
````

````python
print(tf.reshape(Y_test, [10]), tf.reshape(y_preds_4, [10]))
mae_4 = mae(Y_test, y_preds_4) # very good fit
mse_4 = mse(Y_test, y_preds_4)
````
